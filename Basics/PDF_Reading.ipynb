```python
#In PDF file, we can't direct insert any strings. 
#It won't support. As PDF so many parameters. We can append the PDF with another PDF. 
pdf = open("US_Declaration.pdf",mode='rb')
```

```python
pdf_rd = PyPDF2.PdfFileReader(pdf)
```

```python
page_o = pdf_rd.getPage(0)
```

```python
pdf_write = PyPDF2.PdfFileWriter()
```

```python
pdf_write.addPage(page_o)
```

```python
pdf_o = open("Sample.pdf",mode='wb')
```

```python
pdf_write.write(pdf_o)
```

```python
pdf_o.close()
pdf.close()
```
